# Defining Dataset

```python
# Read CSV
import pandas as pd
```

```python
df=pd.read_csv('weather.csv')
```

```python
df.head()
```

# Encoding Features
First, you need to convert these string labels into numbers. for example: 'Overcast', 'Rainy', 'Sunny' as 0, 1, 2. This is known as label encoding. Scikit-learn provides LabelEncoder library for encoding labels with a value between 0 and one less than the number of discrete classes.

```python
weather=df['weather']
temp=df['temp']
play=df['play']
```

```python
# Import LabelEncoder
from sklearn import preprocessing

#creating labelEncoder
le = preprocessing.LabelEncoder()
# Converting string labels into numbers.
df['weather_encoded']=le.fit_transform(weather)
df['temp_encoded']=le.fit_transform(temp)
df['Play_encoded']=le.fit_transform(play)
# print ("Weather:",weather_encoded)
# print ("Temp   :",temp_encoded)
# print ("Play   :",label)

df.head()
```

```python
df_encoded = df[['weather_encoded', 'temp_encoded']].copy()
df_encoded
```

# Generating Model
<!-- Generate a model using naive bayes classifier in the following steps:

```python
#Import Gaussian Naive Bayes model
from sklearn.naive_bayes import GaussianNB

#Create a Gaussian Classifier
model = GaussianNB()
# data= list(features)
# features.reshape(-1, 1)
# print(data)
# Train the model using the training sets
model.fit(df_encoded,df['Play_encoded'])

#Predict Output
predicted= model.predict([[0,2]]) # 0:Overcast, 2:Mild
print ("Predicted Value:", predicted)
```
